# Feature Selection

```python
# Visualizing correlation
plt.figure(figsize=(10,10))
mask=np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)]=True
sns.heatmap(df.corr(),vmin=-1,vmax=1,mask=mask);
```

```python
# Highlight correlation with target variable prog_421a
df.corr()['421a'].sort_values()
```

```python
# Visualize correlation
correlation=df.corr()
sns.heatmap(correlation[['421a']].sort_values('421a',ascending=False), vmin= -1, vmax=1, annot=True);
```

```python
# From correlation information above, construct lists for features with high and low correlation to target variable 421a
# Based on target cutoff of .05 - leaving us with 18 high correlation features to investigate further
high_correlation_features=[]
low_correlation_features=[]

for i in range(len(list(df.corr()['421a']))):
    if abs(list(df.corr()['421a'])[i])>.05:
        high_correlation_features.append(list(df.corr()['421a'].index)[i])
    else:
        low_correlation_features.append(list(df.corr()['421a'].index)[i])

```

```python
# Confirm two lists capture all columns
len(list(df.corr()['421a'])),len(high_correlation_features),len(low_correlation_features)
```

```python
# Re visualize correlation with 421a
# Visualize correlation
correlation=df[high_correlation_features].corr()
sns.heatmap(correlation[['421a']].sort_values('421a',ascending=False), vmin= -1, vmax=1, annot=True);
```

```python
# Drop features from high_correlation_features list that do not align with problem statement
# Features included in data frame for EDA and observation identification purposes
high_correlation_features.remove('421a')
high_correlation_features.remove('bbl')
high_correlation_features.remove('latitude')
high_correlation_features.remove('longitude')

```

```python
# Check for collinearity between high correlation features 
df[high_correlation_features].corr()
```

```python
sns.pairplot(df[high_correlation_features])
```

```python
# Set loop to print columns with high correlation
for col in high_correlation_features:
    for i in range(len(list(df.corr()[col]))):
        if abs(list(df.corr()[col])[i])>.65:
            counter_column=list(df.corr()[col].index)[i]
            print(f'High colinearity between {col} and {counter_column}.')
```
